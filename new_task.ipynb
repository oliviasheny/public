**Female Migrant Ratio**

```python
import pandas as pd
df_female = pd.read_excel ('Sources/UN_MigrantStockByOriginAndDestination_2019.xlsx', sheet_name='Table 3',header=15)
# Table 3 contains the number of female migrants
```

```python
# Rename columns 
df_female=df_female.rename(columns = {'Unnamed: 0': 'Year','Unnamed: 2':'Region of destination'}, inplace = False)

#create a new data frame (df1) that only keeps columns with each African countries
df1_female=df_female[['Year','Region of destination',
       "Algeria", "Angola", "Benin", "Botswana", "Burkina Faso", "Burundi", "Cameroon", "Cabo Verde", 
        "Central African Republic", "Chad", "Comoros", "Congo", "Democratic Republic of the Congo",'Côte d\'Ivoire',
        "Djibouti", "Equatorial Guinea", "Egypt", "Eritrea", "Eswatini","Ethiopia", "Gabon", "Gambia", "Ghana",
        "Guinea", "Guinea-Bissau", "Kenya", "Lesotho", "Liberia", "Libya", "Madagascar","Malawi", "Mali",
        "Mauritania", "Mauritius", "Morocco", "Mozambique", "Namibia", "Niger", "Nigeria","Rwanda","Réunion",
        "Sao Tome and Principe","Saint Helena", "Senegal", "Seychelles","Sierra Leone", "Somalia","South Africa", 
        "South Sudan", "Sudan", "Eswatini", "United Republic of Tanzania", "Togo", "Tunisia", "Uganda",
        "Zambia", "Zimbabwe"]]


#  UN does not recognize Saharawi Arab Democratic Republic

# See the overview of the dataset
df1_female
```

```python
# Create a new dataframe (df2) that only contains the rows where region of destination is the whole world
df2_female=df1_female[df1_female['Region of destination']=='WORLD']

# overview of the new dataframe
df2_female
```

```python
# Create a new column that sums the number of female migrants from all the African countries
df3_female=df2_female.copy()
df3_female['Africa_female']=df2_female.iloc[:,2:].sum(1)

# Clean it a littlbe bit to only contain the total number of female migrants from all African countries
df3_female=df3_female[['Year',"Africa_female"]]
df3_female
```
